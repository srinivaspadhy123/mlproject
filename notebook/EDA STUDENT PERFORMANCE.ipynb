```python
numeric_features = [feature for feature in df.columns if df[feature].dtype !='O']
categorical_feature = [feature for feature in df.columns if df[feature].dtype =='O']

print('We have {} numerical features : {}'.format(len(numeric_features),numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_feature),categorical_feature))
```

```python
# Adding a New Column Total Score and Average Score
```

```python
df['total_score'] = df['math_score']+df['reading_score']+df['writing_score']
df['average_score'] = df['total_score']/3
df.head()
```

```python
reading_full = df[df['reading_score']==100]['average_score'].count()
writing_full = df[df['writing_score'] == 100]['average_score'].count()
math_full = df[df['math_score']==100]['math_score'].count()
```

```python
# Histograms & KDE
```

```python
fig,axs = plt.subplots(1,2,figsize=(15,7))

plt.subplot(121)
sns.histplot(x='average_score',data=df,bins=30,kde=True,color='g')
plt.subplot(122)
sns.histplot(x='average_score',data=df,bins=30,kde=True,hue='gender')
plt.show()
```

```python
fig,axs = plt.subplots(1,2,figsize=(15,7))

plt.subplot(121)
sns.histplot(x='total_score',data=df,bins=30,kde=True,color='g')
plt.subplot(122)
sns.histplot(x='total_score',data=df,bins=30,kde=True,hue='gender')
plt.show()
```

```python
fig,axs = plt.subplots(1,3,figsize=(25,6))

plt.subplot(131)
sns.histplot(x='average_score',data=df,bins=30,kde=True,hue='lunch')
plt.subplot(132)
sns.histplot(x='average_score',data=df[df['gender']=='female'],bins=30,kde=True,hue='lunch')
plt.subplot(133)
sns.histplot(x='average_score',data=df[df['gender']=='male'],bins=30,kde=True,hue='lunch')


plt.show()
```
